```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "/content/drive/MyDrive/Colab Notebooks/Internship/SkillCraft Technology/datasets/Population of India.csv"
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
# Data Cleaning: Rename columns for better readability
df.rename(columns={
    'Population[50]': 'Population',
    'Percent (%)': 'Population_Percent',
    'Male': 'Male_Population',
    'Female': 'Female_Population',
    'Sex ratio': 'Sex_Ratio',
    'Rural[51]': 'Rural_Population',
    'Urban[51]': 'Urban_Population',
    'Area[52] (km2)': 'Area_km2',
    'Density (per km2)': 'Density_per_km2'
}, inplace=True)
```

```python
df.columns
```

```python
# Bar Chart: Male vs Female Population by State
plt.figure(figsize=(12,6))
df_sorted = df.sort_values(by='Population', ascending=False)
plt.barh(df_sorted['State/UT'], df_sorted['Male_Population'], label='Male', color='blue')
plt.barh(df_sorted['State/UT'], df_sorted['Female_Population'], left=df_sorted['Male_Population'], label='Female', color='pink')
plt.xlabel("Population")
plt.ylabel("State/UT")
plt.title("Male vs Female Population by State")
plt.legend()
plt.show()
```

```python

# Histogram: Population Density Distribution
plt.figure(figsize=(8,6))
sns.histplot(df['Density_per_km2'], bins=10, kde=True, color='purple')
plt.xlabel("Density (per km2)")
plt.ylabel("Frequency")
plt.title("Distribution of Population Density")
plt.show()
```

```python
# Pie Chart: Rural vs Urban Population
total_rural = df['Rural_Population'].sum()
total_urban = df['Urban_Population'].sum()
labels = ['Rural', 'Urban']
sizes = [total_rural, total_urban]
colors = ['green', 'orange']
plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140)
plt.title("Rural vs Urban Population Distribution")
plt.show()
```

```python
# Insights:
print("\nKey Insights:")
print("1. Uttar Pradesh has the highest population, while smaller states have significantly lower populations.")
print("2. Most states have a sex ratio below 1000, indicating a male-dominated population.")
print("3. The majority of India's population resides in rural areas, as shown in the pie chart.")
```
