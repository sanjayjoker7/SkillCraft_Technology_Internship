```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = "/content/drive/MyDrive/Colab Notebooks/Internship/SkillCraft Technology/datasets/Titanic-Dataset.csv"  # Make sure the path is correct
df = pd.read_csv(file_path)
```

```python
# Display basic information about the dataset
print("Dataset Overview:")
print(df.info())  # Check data types and missing values
print(df.head())  # Show first few rows
```

```python
# STEP 1: Handling Missing Values
print("\nMissing Values:\n", df.isnull().sum())
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
# Drop Cabin column (too many missing values)
df.drop(columns=['Cabin'], inplace=True)
```

```python
# STEP 2: Data Exploration & Summary Statistics
print("\nSummary Statistics:\n", df.describe())
```

```python
# Countplot for survival
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Survived', hue='Survived', palette='coolwarm', legend=False)
plt.title("Survival Count")
plt.show()
```

```python
# STEP 3: Visualizing Data
# Distribution of Age
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], bins=30, kde=True, color='blue')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# Survival rate by gender
plt.figure(figsize=(6, 4))
sns.barplot(data=df, x='Sex', y='Survived', hue='Sex', palette='pastel', legend=False)
plt.title("Survival Rate by Gender")
plt.show()
```

```python
# Survival rate by class
plt.figure(figsize=(6, 4))
sns.barplot(data=df, x='Pclass', y='Survived', hue='Pclass', palette='muted', legend=False)
plt.title("Survival Rate by Passenger Class")
plt.show()
```

```python
# Pairplot of numerical features
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'Fare']], hue='Survived', palette='husl')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 6))
# Select only numerical features for correlation calculation
numerical_df = df.select_dtypes(include=np.number)
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm', linewidths=1)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# STEP 4: Insights & Observations
print("\nInsights from EDA:")
print("- Higher survival rates for women than men.")
print("- Passengers in 1st class had a higher survival rate than those in 3rd class.")
print("- Younger passengers had a slightly better chance of survival.")

```
